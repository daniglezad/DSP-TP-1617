```python
def dirac(n=0, N=40):
    # Check that the parameter are valid
    if n > N - 1:
        raise ValueError('n should be inferior to N-1')
    
    # Create the array
    x = np.zeros((N, ))
    # Place the dirac
    x[n] = 1
    
    return x

x=dirac(20)
plt.stem(x)
plt.xlabel(r'$n$')
plt.ylabel(r'$\delta(k-n)$')
plt.title('Dirac function')
```

**2.2**

```python
alpha=1
s=1 #as Ts is 1
a=np.exp(-1)

yc = np.zeros((40,))
yn = np.zeros((40,))

for i in range (2, 40):
    yc[i] = x[i] + a*(alpha - 1)*x[i-1]  + 2*a*yc[i-1] - (a**2)*yc[i-2]

plt.figure()
plt.title('causal part')
plt.stem(yc)    

for i in range (37,-1, -1):
    yn[i] = a*(alpha + 1)* x[i+1] - (a**2)*x[i+2] + 2*a*yn[i+1] - (a**2)*yn[i+2]

plt.figure()
plt.title('anticausal part')
plt.stem(yn)    

yt=yc+yn
plt.figure()
plt.title('both')
plt.stem(yt)    
```

```python
def step(n=0, N=20):
    # Check that the parameter are valid
    if n > N - 1:
        raise ValueError('n should be inferior to N-1')
    
    # Create the array
    s = np.zeros((N, ))
    # Create the ramp
    for idx_s in np.arange(n, N):
        s[idx_s] = 1
    
    return s

N = 40
## Step function H(k-4)
x = step(10,N) - step(30,N)
plt.figure
plt.stem(x)
```

```python
#first horizontal

def gradient_hor (Im):
    [nrow, ncol] = np.shape(Im)
    Im_s = np.zeros((nrow,ncol))
    Im_s_c = np.zeros((nrow,ncol))
    Im_s_n = np.zeros((nrow,ncol))
    Im_d_c = np.zeros((nrow,ncol))
    Im_d_n = np.zeros((nrow,ncol))


    #smoothing on columns, derivative on rows
    alpha=1
    s=1 #as Ts is 1
    a=np.exp(-1)
    
    for i in range (2, ncol):
        Im_s_c[i,:] = Im[i,:] + a*(alpha - 1)*Im[i-1,:]  + 2*a*Im_s_c[i-1,:] - (a**2)*Im_s_c[i-2,:]
    for i in range (Im.shape[0]-3,-1, -1):
        Im_s_n[i,:] = a*(alpha + 1)* Im[i+1,:] - (a**2)*Im[i+2,:] + 2*a*Im_s_n[i+1,:] - (a**2)*Im_s_n[i+2,:]

    Im_smoothed = Im_s_c + Im_s_n

    
    #now derivative of the already smoothed columns
    for i in range (2, nrow):
        Im_d_c[:,i] = -s*alpha*a*Im_smoothed[:,i-1]   + 2*a*Im_d_c[:,i-1] - (a**2)*Im_d_c[:,i-2]
    for i in range (Im.shape[0]-3,-1, -1):
        Im_d_n[:,i] = s*alpha*a*Im_smoothed[:,i+1]   + 2*a*Im_d_n[:,i+1] - (a**2)*Im_d_n[:,i+2]

    Im_derivated = Im_d_c + Im_d_n
    
    return Im_derivated

x = gradient_hor (dali_im)
plt.figure()
plt.title('horizontal')
io.imshow(x, cmap=plt.cm.gray)

#now the vertical

def gradient_vert (Im):
    [nrow, ncol] = np.shape(Im)
    Im_s = np.zeros((nrow,ncol))
    Im_s_c = np.zeros((nrow,ncol))
    Im_s_n = np.zeros((nrow,ncol))
    Im_d_c = np.zeros((nrow,ncol))
    Im_d_n = np.zeros((nrow,ncol))


    #smoothing on rows, derivative on columns
    alpha=1
    s=1 #as Ts is 1
    a=np.exp(-1)
    
    for i in range (2, ncol):
        Im_s_c[:,i] = Im[:,i] + a*(alpha - 1)*Im[:,i-1]  + 2*a*Im_s_c[:,i-1] - (a**2)*Im_s_c[:,i-2]
    for i in range (Im.shape[0]-3,-1, -1):
        Im_s_n[:,i] = a*(alpha + 1)* Im[:,i+1] - (a**2)*Im[:,i+2] + 2*a*Im_s_n[:,i+1] - (a**2)*Im_s_n[:,i+2]

    Im_smoothed = Im_s_c + Im_s_n

    
    #now derivative of the already smoothed rows
    for i in range (2, nrow):
        Im_d_c[i,:] = -s*alpha*a*Im_smoothed[i-1,:]   + 2*a*Im_d_c[i-1,:] - (a**2)*Im_d_c[i-2,:]
        
    for i in range (Im.shape[0]-3,-1, -1):
        Im_d_n[i,:] = s*alpha*a*Im_smoothed[i+1,:]   + 2*a*Im_d_n[i+1,:] - (a**2)*Im_d_n[i+2,:]

    Im_derivated = Im_d_c + Im_d_n
    
    return Im_derivated


y = gradient_vert (dali_im)
plt.figure()
plt.title('vertical')
io.imshow(y, cmap=plt.cm.gray)

#smoothing on rows, derivative on columns
```
